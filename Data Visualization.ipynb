```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('hcare.csv')
```

```python
data['Hospital_code'] = data['Hospital_code'].astype(object)
data['City_Code_Hospital'] = data['City_Code_Hospital'].astype(object)
# data['Available Extra Rooms in Hospital'] = data['Available Extra Rooms in Hospital'].astype(object)
data['Bed Grade'] = data['Bed Grade'].astype(object)
data['City_Code_Patient'] = data['City_Code_Patient'].astype(object)
```

```python
cat_cols = list(data.select_dtypes(include=['object']).columns)
for col in cat_cols:
    count = data[col].value_counts()
    plt.figure()
    sns.barplot(x=count.index.values, y=count.values,data=data)
    plt.xlabel(col)
    plt.ylabel('No of occurences')
```

```python
num_cols = list(data.select_dtypes(include=['int64','float64']).columns)
for col in num_cols:
    sns.displot(data[col], kde=True)
```

```python
for i in cat_cols:

            plt.figure(figsize=(10,5))

            sns.countplot(x='Stay',hue=i,data=data)
```

```python
from sklearn.preprocessing import LabelEncoder
cat_col = list(data.select_dtypes(include=['object']).columns)

label = LabelEncoder()
for col in cat_col:
    data[col] = label.fit_transform(data[col])
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(ax=ax, data=data.corr(), cmap="crest", annot=True, cbar=True).set_title('Correlation Plot of LOS dataset',fontdict= { 'fontsize': 30})
```
